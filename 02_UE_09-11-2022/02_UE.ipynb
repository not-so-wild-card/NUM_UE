```octave
# backSubst kann Gleichungssysteme mit unendlich vielen Lösungen nicht lösen (Rang(U) muss vollen Rang haben)
function ret = backSubst (b, U)
    if rows(U) != columns(U)
        error ("Ist nicht eine obere Dreiecksmatrix!");
        return
    endif
    if rows(b) != columns(U)
        error ("Anzahl der Vektorelemente passt nicht!");
        return
    endif
    ret = b;
    for i = rows(U):-1:1
        for j = columns(U):-1:i
            if j == i
                if U(i,j) == 0
                    error ("Achte darauf, dass die Matrix vollen Rang hat!"); #?!
                    return
                endif
                ret(i) = ret(i)/U(i,j);
            else
                ret(i) = ret(i)-ret(j)*U(i,j);
            endif
        endfor
    endfor
endfunction
```

```octave
b1 = [0; -4; 0];
 
U1 = [2, 0, 30;
      0, 4, 0; 
      0, 0, 6];

meineLoesung1 = backSubst(b1, U1)
disp("\n")
check1 = U1\b1
```
